# Deteccion de Bordes - PDI

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Carga de imagen

```python
ruta_imagen = "imagen.jpg"
imagen_color = cv2.imread(ruta_imagen)
if imagen_color is None:
    raise ValueError(f"No se pudo cargar: {ruta_imagen}")
plt.imshow(cv2.cvtColor(imagen_color, cv2.COLOR_BGR2RGB))
plt.title("Original")
plt.axis("off")
plt.show()
```

## 2. Preprocesamiento

```python
imagen_gris = cv2.cvtColor(imagen_color, cv2.COLOR_BGR2GRAY)
imagen_suavizada = cv2.GaussianBlur(imagen_gris, (5, 5), 0)
plt.imshow(imagen_gris, cmap='gray')
plt.title("Escala de Grises")
plt.axis("off")
plt.show()
```

## 3. Operadores de Gradiente - Implementacion Manual

```python
# SOBEL
sobel_x = cv2.Sobel(imagen_suavizada, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(imagen_suavizada, cv2.CV_64F, 0, 1, ksize=3)
sobel_mag = np.sqrt(sobel_x**2 + sobel_y**2)
sobel_mag = np.uint8(sobel_mag / sobel_mag.max() * 255)
plt.imshow(sobel_mag, cmap='gray')
plt.title("Sobel")
plt.axis("off")
plt.show()
```

```python
# PREWITT
kernel_prewitt_x = np.array([[-1, 0, 1],
                             [-1, 0, 1],
                             [-1, 0, 1]], dtype=np.float32)
kernel_prewitt_y = np.array([[-1, -1, -1],
                             [0, 0, 0],
                             [1, 1, 1]], dtype=np.float32)
prewitt_x = cv2.filter2D(imagen_suavizada, cv2.CV_64F, kernel_prewitt_x)
prewitt_y = cv2.filter2D(imagen_suavizada, cv2.CV_64F, kernel_prewitt_y)
prewitt_mag = np.sqrt(prewitt_x**2 + prewitt_y**2)
prewitt_mag = np.uint8(prewitt_mag / prewitt_mag.max() * 255)
plt.imshow(prewitt_mag, cmap='gray')
plt.title("Prewitt")
plt.axis("off")
plt.show()
```

```python
# ROBERTS
kernel_roberts_x = np.array([[1, 0],
                            [0, -1]], dtype=np.float32)
kernel_roberts_y = np.array([[0, 1],
                            [-1, 0]], dtype=np.float32)
roberts_x = cv2.filter2D(imagen_suavizada, cv2.CV_64F, kernel_roberts_x)
roberts_y = cv2.filter2D(imagen_suavizada, cv2.CV_64F, kernel_roberts_y)
roberts_mag = np.sqrt(roberts_x**2 + roberts_y**2)
roberts_mag = np.uint8(roberts_mag / roberts_mag.max() * 255)
plt.imshow(roberts_mag, cmap='gray')
plt.title("Roberts")
plt.axis("off")
plt.show()
```

```python
# LAPLACIANO
laplaciano = cv2.Laplacian(imagen_suavizada, cv2.CV_64F, ksize=3)
laplaciano_img = np.abs(laplaciano)
laplaciano_img = np.uint8(laplaciano_img / laplaciano_img.max() * 255)
plt.imshow(laplaciano_img, cmap='gray')
plt.title("Laplaciano")
plt.axis("off")
plt.show()
```

## 4. Canny (cv2)

```python
canny_img = cv2.Canny(imagen_suavizada, 50, 150)
plt.imshow(canny_img, cmap='gray')
plt.title("Canny")
plt.axis("off")
plt.show()
```

## 5. Comparacion General

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes[0, 0].imshow(cv2.cvtColor(imagen_color, cv2.COLOR_BGR2RGB))
axes[0, 0].set_title("Original")
axes[0, 0].axis("off")

axes[0, 1].imshow(imagen_gris, cmap='gray')
axes[0, 1].set_title("Escala de Grises")
axes[0, 1].axis("off")

axes[0, 2].imshow(sobel_mag, cmap='gray')
axes[0, 2].set_title("Sobel")
axes[0, 2].axis("off")

axes[1, 0].imshow(prewitt_mag, cmap='gray')
axes[1, 0].set_title("Prewitt")
axes[1, 0].axis("off")

axes[1, 1].imshow(roberts_mag, cmap='gray')
axes[1, 1].set_title("Roberts")
axes[1, 1].axis("off")

axes[1, 2].imshow(canny_img, cmap='gray')
axes[1, 2].set_title("Canny")
axes[1, 2].axis("off")

plt.tight_layout()
plt.show()
```
