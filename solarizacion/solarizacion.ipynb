```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def solarizar(imagen,umbral):
    #aplicar solarizacion
    img_sol = np.where(imagen < umbral, imagen, 255 - imagen)
    return img_sol
imagen = cv2.imread('imagen.jpg')
img_grey = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
imgFinal = solarizar(img_grey, 128)
plt.imshow(imgFinal, cmap='gray')
```
